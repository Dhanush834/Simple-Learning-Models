```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0,2.0]) # x denotes Size of house in sq.feet & are in 1000sq.feet Scaling
y_train = np.array([300.0,500.0]) # y denotes Price of house in dollars & are in 1000s of $
```

```python
#Ploting the data set 
plt.scatter(x_train,y_train,marker='x',c='red')
plt.show()
```

```python
# fit a linear regression st.line for the above data f = w * x + b
# trying the best values of w and b
# THIS VALUE IS TESTED FOR MIN COST ERROR VALUE AND ALSO GRADIENT DESCENT
w = 200
b = 100

def compute_price (x,w,b):
    m = x.shape[0]
    f = np.zeros(m)
    for i in range(m):
        f[i] = w * x[i] + b
    return f

output = compute_price(x_train,w,b)

#plot Actual Data 
plt.scatter(x_train,y_train,c='r',marker='x')

#plot the model output Predection
plt.plot(x_train, output, c='b',label='Our Prediction')

plt.title("Housing Prices")

plt.ylabel('Price (in 1000s of dollars)')

plt.xlabel('Size (1000 sqft)')

plt.legend(["Actual Values","Our Predection"])

plt.show()
```

```python
def House_price (inp):
    return (200*inp + 100)
    
inp = float(input("Enter Size of House : "))           
cost = House_price(inp)
print(f"The Cost of House for Size {inp} is {cost}K $")
```
